```python
# problem 1: 1.	Compute the longest run of days where usage=0 for each user (longest_no_usage)

def cal_longest_usage(x):
    '''
    Compute the longest run of days where usage=0 for each user (longest_no_usage)
    
    args:
    ----------------------
    x (df): a single user's df with features user, day, usage
    
    returns:
    ----------------------
    longest_no_usage (int): the max count for this user
    '''
    if x['usage'].iloc[0] == 0:
        count = [1] * x.shape[0]
    else:
        count = [0]* x.shape[0]
    for i,j in enumerate(x['usage']):
        if i == 0:
            continue
        if j == 0:
            count[i] = count[i-1]+1
        else:
            count[i] = 0
    longest_no_usage = max(count)
    return longest_no_usage


count_no_usuage = []
for i in usage_df['user'].unique():
    count_no_usuage.append([i, cal_longest_usage(usage_df[usage_df['user'] == i])])
```

```python
count_no_usuage_df = pd.DataFrame(count_no_usuage,columns=['user', 'longest_no_usage'])
```

```python
# Problem 1.a.summarize the results visually by histogram
# conclusion: we can find that the histogram is very left skewed, which indicates that most 
# user's longest usage tend to 0

plt.hist(count_no_usuage_df['longest_no_usage'])
```

```python
# Problem 2: 
# select users with more than 8 days data

# data cleaning
# find user's with more than 8 records and get the index
cond_8 = usage_df.groupby(['user'])['user'].count() >8
user_mt_8dy = usage_df.groupby(['user'])['user'].count()[cond_8].index

# filter user's with more than 8 records
usage_df_mt8 = usage_df[usage_df['user'].isin(user_mt_8dy)]

# transform the 8 days to df and append user id
usage_df_d8 = pd.DataFrame(usage_df_mt8.groupby('user').head(8)['usage'].values.reshape(-1,8), columns = ['d' + str(i) for i in range(1,9)])
usage_df_d8['user'] = usage_df_mt8['user'].unique()

# merge with user's age and gender
usage_df_d8 = usage_df_d8.merge(user_df, on = 'user')
```

```python
usage_df_d8.head()
```

```python
# train-test split and further split to LSTM and MLP's input
x,y = usage_df_d8.loc[:, usage_df_d8.columns != 'd8'],usage_df_d8['d8']
train_X, test_X, train_y, test_y = sklearn.model_selection.train_test_split(x,y,test_size= 0.2, stratify = y)
train_X_LSTM = train_X[['d1','d2','d3','d4','d5','d6','d7']].values
train_X_LSTM = train_X_LSTM.reshape((train_X_LSTM.shape[0],train_X_LSTM.shape[1],1))
train_X_MLP = train_X[['age','gender']].values


test_X_LSTM = test_X[['d1','d2','d3','d4','d5','d6','d7']].values
test_X_LSTM = test_X_LSTM.reshape((test_X_LSTM.shape[0],test_X_LSTM.shape[1],1))
test_X_MLP = test_X[['age','gender']].values


train_y, test_y = train_y.values, test_y.values
```

```python
from tensorflow.keras.layers import Input, LSTM, Dense, Concatenate
from tensorflow.keras.models import Model

#lstm input shape
lstm_input = Input(shape=train_X_LSTM.shape[1:])

#lstm units
hidden1 = LSTM(20)(lstm_input)

#lstm output 
lstm_output = Dense(5, activation='sigmoid')(hidden1)

#mlp input with additonal 2 variables 
mlp_input=Input(shape=(train_X_MLP.shape[1]))

#combine the lstm output which is predicted y
x = Concatenate()([lstm_output, mlp_input])
x = Dense(5,activation='relu' )(x)

#mlp model output which is predicted y
mlp_out = Dense(1, activation='sigmoid')(x)

#final output of combined model which is predicted y
model = Model(inputs=[lstm_input, mlp_input],outputs=mlp_out)
```

```python
# view the model
model.summary()
```

```python
#compile the model
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

# #fit the model
model.fit([train_X_LSTM, train_X_MLP], train_y, batch_size=64, epochs=50, validation_split=0.2)
```

```python
# test result

((model.predict([test_X_LSTM, test_X_MLP]) > 0.5).reshape(-1).astype('int') == test_y).sum()/test_y.shape[0]
```
